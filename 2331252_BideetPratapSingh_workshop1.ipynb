```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Load a grayscale image
image = Image.open('/content/drive/MyDrive/Artificial Intelligence and Machine Learning/camera_man.jpg')
image_gray = image.convert("L")  # Convert the image to grayscale

# 1. Display the grayscale image
plt.figure(figsize=(6, 6))
plt.imshow(image_gray, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')  # Hide axes for cleaner display
plt.show()
```

```python
# 2. Extract and display the middle section of the image (150 pixels)
image_array = np.array(image_gray)

# Find the center of the image
height, width = image_array.shape
center_x, center_y = width // 2, height // 2

# Define the size of the cropped section (150x150 pixels)
crop_size = 150
start_x = center_x - crop_size // 2
start_y = center_y - crop_size // 2

# Extract the middle 150x150 section
cropped_image = image_array[start_y:start_y+crop_size, start_x:start_x+crop_size]

# Display the cropped image
plt.figure(figsize=(6, 6))
plt.imshow(cropped_image, cmap='gray')
plt.title('Cropped 150x150 Section')
plt.axis('off')  # Hide axes
plt.show()
```

```python
# 3. Apply a simple threshold to the image
# Set all pixel values below 100 to 0 and all values above 100 to 255 (binary thresholding)
thresholded_image = np.where(image_array < 100, 100, 255)

# Display the thresholded binary image
plt.figure(figsize=(6, 6))
plt.imshow(thresholded_image, cmap='gray')
plt.title('Thresholded Binary Image')
plt.axis('off')
plt.show()
```

```python
# 4. Rotate the image 90 degrees clockwise
rotated_image = image_gray.rotate(270)  # 270 degrees = 90 degrees clockwise

# Display the rotated image
plt.figure(figsize=(6, 6))
plt.imshow(rotated_image, cmap='gray')
plt.title('Rotated Image (90 degrees clockwise)')
plt.axis('off')
plt.show()
```

```python
# 5. Convert the grayscale image to an RGB image
rgb_image = np.stack([image_array, image_array, image_array], axis=-1)

# Display the converted RGB image
plt.figure(figsize=(6, 6))
plt.imshow(rgb_image)
plt.title('Grayscale to RGB Image')
plt.axis('off')
plt.show()
```

# Exercise 3

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

image = Image.open('/content/drive/MyDrive/Artificial Intelligence and Machine Learning/camera_man.jpg').convert("L")

image_array = np.array(image)
print(image_array.shape)

height, width = image_array.shape

# data = image_array.reshape(height, width)
data = image_array.copy()

plt.imshow(image_array, cmap='gray')
plt.show()
```

```python
# Compute the mean of each column (feature)
mean = np.mean(data, axis=0)

# Subtract mean to center the data
centered_data = data - mean
centered_data = centered_data
```

```python
centered_data.shape
```

```python
cov_matrix = np.cov(centered_data, rowvar=False)
```

```python
cov_matrix.shape
```

```python
eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)

sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]
```

```python
explained_varience_ratio = eigenvalues / np.sum(eigenvalues)
plt.plot(np.cumsum(explained_varience_ratio))
plt.title("Cumulative Explained Varience Ratio")
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Varience Ratio")
plt.grid(True)
plt.show()
```

```python
k = 50
components = eigenvectors [: , : k]
```

```python
compressed_data = np.dot(centered_data, components)
```

```python
decompressed_data = np.dot(compressed_data, components.T) + mean
```

```python
plt.figure(figsize=(12, 6))
#Original Image
plt.subplot(1, 3, 1)
plt.imshow(image_array, cmap='gray')
plt.title("Original Image")
plt.axis("off")
#Compressed Image
plt.subplot(1, 3, 2)
plt.imshow(compressed_data, cmap = "gray", aspect = "auto")
plt.title(f"Compressed Image (k = {k})")
plt.axis("off")
#Decompressed Image
plt.subplot(1, 3, 3)
plt.imshow(decompressed_data, cmap = "gray")
plt.title("Decompressed Image")
plt.axis("off")
plt.show()
```
